```python
import matplotlib.pyplot as plt
with open("triplet_esm2_eval.out") as f:
    result_35m = f.readlines()
```

```python
result_35m_precisions = []
result_35m_recalls = []
result_35m_F1s = []
result_35m_AUCs = []
for line in result_35m:
    if line.startswith(">>> pre"):
        result_35m_precisions.append(float(line.split("precision: ")[1].split("|")[0]))
        result_35m_recalls.append(float(line.split("recall: ")[1].split("|")[0]))
        result_35m_F1s.append(float(line.split("F1: ")[1].split("|")[0]))
        result_35m_AUCs.append(float(line.split("AUC: ")[1]))
```

```python
import matplotlib.pyplot as plt
with open("triplet_esm1_600m_eval.out") as f:
    result_650m = f.readlines()


result_650m_precisions = []
result_650m_recalls = []
result_650m_F1s = []
result_650m_AUCs = []
for line in result_650m:
    if line.startswith(">>> pre"):
        result_650m_precisions.append(float(line.split("precision: ")[1].split("|")[0]))
        result_650m_recalls.append(float(line.split("recall: ")[1].split("|")[0]))
        result_650m_F1s.append(float(line.split("F1: ")[1].split("|")[0]))
        result_650m_AUCs.append(float(line.split("AUC: ")[1]))
```

```python
plt.plot(range(50, 1000, 50), result_35m_precisions, label='esm2 35M')
plt.plot(range(50, 1000, 50), result_150m_precisions, label='esm2 150M')
plt.plot(range(50, 1000, 50), result_650m_precisions, label='esm1 650M')
plt.legend()
plt.title("Precision")
```

```python
plt.plot(range(50, 1000, 50), result_35m_recalls, label='esm2 35M')
plt.plot(range(50, 1000, 50), result_150m_recalls, label='esm2 150M')
plt.plot(range(50, 1000, 50), result_650m_recalls, label='esm1 650M')
plt.legend()
plt.title("Recall")
```

```python
plt.plot(range(50, 1000, 50), result_35m_F1s, label='esm2 35M')
plt.plot(range(50, 1000, 50), result_150m_F1s, label='esm2 150M')
plt.plot(range(50, 1000, 50), result_650m_F1s, label='esm1 650M')
plt.legend()
plt.title("F1")
```

```python
plt.plot(range(50, 1000, 50), result_35m_AUCs, label='esm2 35M')
plt.plot(range(50, 1000, 50), result_150m_AUCs, label='esm2 150M')
plt.plot(range(50, 1000, 50), result_650m_AUCs, label='esm1 650M')
plt.legend()
plt.title("AUC")
```

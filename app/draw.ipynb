```python
import matplotlib.pyplot as plt
with open("triplet_esm2_eval.out") as f:
    contents = f.readlines()
```

```python
precisions = []
recalls = []
F1s = []
AUCs = []
for line in contents:
    if line.startswith(">>> pre"):
        precisions.append(float(line.split("precision: ")[1].split("|")[0]))
        recalls.append(float(line.split("recall: ")[1].split("|")[0]))
        F1s.append(float(line.split("F1: ")[1].split("|")[0]))
        AUCs.append(float(line.split("AUC: ")[1]))
```

```python
plt.plot(range(50, 1000, 50), precisions, label='precision')
plt.plot(range(50, 1000, 50), recalls, label='recall')
plt.plot(range(50, 1000, 50), F1s, label='F1')
plt.plot(range(50, 1000, 50), AUCs, label='AUC')
plt.legend()
```
